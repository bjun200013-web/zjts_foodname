```python
food_depth_path = r'/date0/crwu/zjts_foodname/data/classification/data/food_depth.csv'
depth_df = read_dataset(food_depth_path)

depth_md = '''| 第一级别 | 第二级别 | 第三级别 | 第四级别 |
| :--- | :--- | :--- | :--- |'''
last_level_1 = ''
last_level_2 = ''
last_level_3 = ''
last_level_4 = ''
for row in depth_df.iterrows():
    level_1, level_2, level_3,level_4 = row[1]
    # print(level_1, level_2, level_3,level_4)
    if level_1 == last_level_1:
        current_level_1 = ' '
    else:
        current_level_1 = f' {level_1} '
        last_level_1 = level_1
    if level_2 == last_level_2:
        current_level_2 = ' '
    else:
        current_level_2 = f' {level_2} '
        last_level_2 = level_2
    if level_3 == last_level_3:
        current_level_3 = ' '
    else:
        current_level_3 = f' {level_3} '
        last_level_3 = level_3
    if level_4 == last_level_4:
        current_level_4 = ' '
    else:
        current_level_4 = f' {level_4} '
        last_level_4 = level_4

    depth_md += f'\n|{current_level_1}|{current_level_2}|{current_level_3}|{current_level_4}|'
print(depth_md)
        
```
